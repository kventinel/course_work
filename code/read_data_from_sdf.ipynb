```python
import gzip
```

```python
atom = False
structure = []
column = -1
i = 0
data = []
diff_atoms = set()
with open('data/BindingDB_All_3D_3.sdf', 'r') as fin:
    lines = fin.readlines()
    for line in lines:
        if column != -1:
            if column == 1:
                data[i][column] = float(line[:-1])
            else:
                data[i][column] = line[:-1].split(',')[0]
            column = -1
        if line == 'M  V30 BEGIN ATOM\n':
            atom = True
        elif line == 'M  V30 END ATOM\n':
            atom = False
            data.append([0, 0, 0, 0])
            data[i][0] = structure
            structure = []
        elif line == '>  <Ki (nM)>\n':
            column = 1
        elif line == '>  <PDB ID(s) for Ligand-Target Complex>\n':
            column = 2
        elif line == '>  <PDB ID(s) of Target Chain>\n':
            column = 3
        elif line == '$$$$\n':
            i += 1
        elif atom:
            split = line[3:-1].split(' ')
            structure.append([split[2], float(split[3]), float(split[4]), float(split[5])])
            diff_atoms.add(split[2])
```

```python
used = []
for line in data:
    if line[2][0] in '12' and line[3][0] in '12':
        used.append(line)
```

```python
k = 0
for line in data:
    try:
        f = gzip.open('data/ftp.rcsb.org/pub/pdb/data/structures/all/pdb/pdb' + line[3].lower() + '.ent.gz', 'rb')
        lines = f.readlines()
        f.close()
        for line in lines:
            line = line.split()
            if line[0] == b'ATOM':
                diff_atoms.add(line[-1])
                k += 1
    except FileNotFoundError:
        pass
print(k)
```
